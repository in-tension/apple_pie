```python
import ipywidgets as wids
import bqplot
from bqplot.colorschemes import CATEGORY10
from IPython.display import display as idisplay
%matplotlib inline
```

```python
def nb_disp_exp(exp) :
    group_tabs = wids.Tab()
    tab_children = []
    for group in exp.groups.values() :
        group_widget = make_group_widget(group)
        tab_children.append(group_widget)
        
    group_tabs.children = tab_children
    
    #i = 0;
    for i,group_name in enumerate(exp.groups.keys()) :
        group_tabs.set_title(i,group_name)
    idisplay(group_tabs)
```

```python
import time
def update_plot(scat, show) :
    scat.visible = show
```

```python
def make_group_widget(group) :
    
    x_sc = bqplot.LinearScale()
    y_sc = bqplot.LinearScale()
    #fig_wid = bqplot.Axis()
#     print(fig_wid.__dict__.keys())
    #axes = plt.axes()
    #print(type(axes))

    vb = wids.VBox()
    vb_children = []
    
#     checkboxes = []
    scats = []
    #output_widget = wids.Output()
    c = 0
    temp_list = [group.exper.control]
    temp_list.extend(list(group.condits.values()))
    for condit in temp_list :
        try :
                  
            condit_cb = wids.Checkbox(
                value=False,
                description = condit.name_str,
                disabled = False
            )

            scat = bqplot.Scatter(
                x=condit.t_int,
                y=condit.dist_means,
                scales={'x':x_sc, 'y':y_sc},
                label = [condit.name_str],
#                 display_legend=True,
                colors=[CATEGORY10[c]],
                visible=False
            )
            print(condit_cb.layout.keys)
            print(condit_cb.keys)

            c += 1
            scats.append(scat)

            intera = wids.interactive(
                update_plot,
                scat=wids.fixed(scat),
                show=condit_cb
            )

            vb_children.append(intera)
        
        except Exception as e:
            #print(e.__dict__.keys())
            print('condit: {}, exception: {}'.format(condit.name,e))
            #e.__traceback__.print_last()
        
    vb.children = vb_children
    
    x_axis_label = 'time'
    y_axis_label = 'displacement'
    
    x_ax = bqplot.Axis(
        label=x_axis_label, 
        scale=x_sc,
        tick_format='0.0f',
        grid_lines='solid',
        orientation='horizontal',)
    
    y_ax = bqplot.Axis(
        label=y_axis_label, 
        scale=y_sc,
        tick_format='0.0f',
        grid_lines='solid',
        orientation='vertical',)

    fig_wid = bqplot.Figure(
        axes=[x_ax, y_ax],  
        marks=scats, 
        legend_location = 'top-right')

    
   
    
#     hb = ipywidgets.HBox((vb,output_widget))
    hb = wids.HBox((vb,fig_wid))
#     hb = wids.HBox((fig_wid,))
    return hb

    
nb_disp_exp(exp)
        

        
```
